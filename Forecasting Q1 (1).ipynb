```python
## calculates the Mean Absolute Percentage Error between predicted and actual values.
```

```python
def MAPE(pred,org):
    temp = np.abs((pred-org)/org)*100
    return np.mean(temp)
```

```python
##calculates the root mean squared error between the original and predicted values.
```

```python
def rmse(org,pred):
    rmse= np.sqrt(np.mean((np.array(org)-np.array(pred))**2))
    return rmse
```

# Linear Model

```python
##fitting a linear regression model to your training data and then making predictions on your test data.Finally calculate and print the RMSE
```

```python
import statsmodels.formula.api as smf
linear_model = smf.ols('Passengers~t', data = train).fit()
pred_linear = pd.Series(linear_model.predict(test['t']))
rmse_linear = rmse(test.Passengers,pred_linear)
rmse_linear
```

# Exponential Model

```python
## fitting an exponential (logarithmic) model and calculate RMSE
```

```python
exp_model = smf.ols('log_passengers~t', data=train).fit()
pred_exp = pd.Series(exp_model.predict(test['t']))
rmse_exp = np.sqrt(np.mean((np.array(test['Passengers'])-np.array(np.exp(pred_exp)))**2))
rmse_exp
```

# Quadratic Model

```python
##fitting a quadratic model and calculate RMSE 
```

```python
quad_model = smf.ols('Passengers~t+t_square', data=train).fit()
pred_quad = pd.Series(quad_model.predict(test[['t','t_square']]))
rmse_quad = np.sqrt(np.mean((np.array(test['Passengers'])-np.array(pred_quad))**2))
rmse_quad
```

# Simple Exponential Method


```python
##fit a SES model to your training data and making predictions on the test set  and calculating both MAPE and RMSE
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
ses_model = SimpleExpSmoothing(train['Passengers']).fit(smoothing_level = 0.2)
pred_ses = ses_model.predict(start = test.index[0],end = test.index[-1])
sme = MAPE(pred_ses,test.Passengers)
print("MAPE is : ", sme)
rmse_sem = rmse(test.Passengers,pred_ses)
rmse_sem
```

# Holt Method


```python
##fit a Holt model to your training data and making predictions on the test set and calculating both MAPE and RMSE
```

```python
from statsmodels.tsa.holtwinters import Holt
hw_model = Holt(train['Passengers']).fit()
pred_hw = hw_model.predict(start = test.index[0],end = test.index[-1])
holt = MAPE(pred_hw,test.Passengers)
holt
rmse_holt = rmse(test.Passengers,pred_hw)
rmse_holt
```

```python
hw_model = Holt(train['Passengers']).fit(smoothing_level = 0.2, smoothing_slope = 0.2)
pred_hw = hw_model.predict(start = test.index[0], end = test.index[-1])
mape_hw = MAPE(pred_hw, test.Passengers)
print("MAPE is : ", mape_hw)
rmse_hw = rmse(test.Passengers, pred_hw)
rmse_hw
```

# Holts winter exponential smoothing with additive seasonality and additive trend

```python
##Exponential Smoothing method fitting a model to your training data and making predictions on the test set and calculating both MAPE and RMSE 
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
hwe_model_add_add = ExponentialSmoothing(train['Passengers'],seasonal = 'add', trend = 'add', seasonal_periods = 4).fit()
pred_hwe = hwe_model_add_add.predict(start=test.index[0], end = test.index[-1])
haa = MAPE(pred_hwe, test.Passengers)
print("MAPE is :",haa)
rmse_haa = rmse(test.Passengers, pred_hwe)
rmse_haa
```

# Holts winter exponential smoothing with multiplicative seasonality and additive trend

```python
##Exponential Smoothing method fitting a model to your training data and making predictions on the test set and calculating both MAPE and RMSE 
```

```python
hwe_model_mul_add = ExponentialSmoothing(train['Passengers'],seasonal = 'mul', trend = 'add', seasonal_periods = 4).fit()
pred_hwe_ma = hwe_model_mul_add.predict(start=test.index[0],end = test.index[-1])
hma = MAPE(pred_hwe_ma,test.Passengers)
print('MAPE is : ',hma)
rmse_hma = rmse(test.Passengers,pred_hwe_ma)
rmse_hma
```
