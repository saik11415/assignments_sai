```python
#Feature Selection by using Mutual Information Feature Selection
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, mutual_info_regression
```

```python
## selects features according to the k highest scores of a certain scoring function
##applies the feature selection method to the training and testing data
```

```python
def select_features(x_train,y_train,x_test):
  fs = SelectKBest(score_func = mutual_info_regression, k ='all')
  fs.fit(x_train, y_train)
  x_train_fs = fs.transform(x_train)
  x_test_fs = fs.transform(x_test)
  return x_train_fs, x_test_fs,fs
```

```python
##suppress or ignore warning messages that may be generated during the execution of your code.
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
##split your data into training and testing sets
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.3, random_state = 24)
x_train_fs, x_test_fs,fs = select_features(x_train, y_train,x_test)
```

```python
##creating a bar plot of the scores associated with each feature in your dataset
```

```python
import matplotlib.pyplot as plt
for i in range(len(fs.scores_)):
  print('Feature %d: %f' % (i, fs.scores_[i]))
fig, ax = plt.subplots(figsize = (10,6))
plt.bar([i for i in range(len(fs.scores_))], fs.scores_)
plt.show()
```

```python
##As per above feature selection method, we will select only features with good score to build our model
```

```python
x = x.drop(['AT','AP','AH','CO','NOX'], axis = 1)
x
```

```python
##split your data into training and testing sets
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)
```

```python
##Artificial Neural Network Model - Backpropagation
```

```python
##Importing Modules
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
##creates a sequential neural network model
```

```python
model = Sequential()
model.add(Dense(8, input_dim = 5, activation = 'relu')) # input layer
model.add(Dense(1,activation = 'sigmoid')) # output layer
```

```python
##configuring the compilation of your Keras model
## mse used as a loss function for regression problems
```

```python
model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics =['mse'])
```

```python
##training your Keras model using the provided training data (x and y) and validate it using the validation split
```

```python
history = model.fit(x, y, validation_split = 0.3, epochs = 250, batch_size = 100)
```

```python
##evaluating your trained model on the same data it was trained 
```

```python
scores = model.evaluate(x,y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
