```python
##categorize the 'Taxable.Income' column into two bins: 'risky' and 'good', based on specified bin edges
```

```python
df['Taxable.Income'] = pd.cut(x=df['Taxable.Income'],bins=[0,30000,100000],labels=['risky','good'],right=False)
Y = df['Taxable.Income']
Y
```

```python
#removing the specified column along the specified axis
```

```python
X = df.drop('Taxable.Income',axis=1)
X
```

```python
#LabelEncoder used to convert categorical labels into numerical values.
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
X['Undergrad'] = LE.fit_transform(X['Undergrad'])
X['Marital.Status'] = LE.fit_transform(X['Marital.Status'])
X['Urban'] = LE.fit_transform(X['Urban'])
X
```

```python
#split your data into training and testing sets
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
# fitting it on the training data to make predictions and evaluate its performance
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
DTC = DecisionTreeClassifier()
DTC.fit(X_train,Y_train)
```

```python
# made predictions on both the training set and the test set and  evaluate the performance of your Decision Tree Classifier
```

```python
Y_pred_train = DTC.predict(X_train)
Y_pred_test = DTC.predict(X_test)
print('Number of nodes is',DTC.tree_.node_count)
print('Maximum depth is',DTC.tree_.max_depth)
print('Training score is',accuracy_score(Y_train,Y_pred_train))
print('Test score is',accuracy_score(Y_test,Y_pred_test))
```

```python
#updating (DTC) by setting the max_depth parameter to 10 and evaluate the performance
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
DTC = DecisionTreeClassifier(max_depth=10)
DTC.fit(X_train,Y_train)
Y_pred_train = DTC.predict(X_train)
Y_pred_test = DTC.predict(X_test)
print('Number of nodes is',DTC.tree_.node_count)
print('Maximum depth is',DTC.tree_.max_depth)
print('Training score is',accuracy_score(Y_train,Y_pred_train))
print('Test score is',accuracy_score(Y_test,Y_pred_test))
```

```python
#updating (DTC) by setting the max_depth parameter to 8 and evaluate the performance
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
DTC = DecisionTreeClassifier(criterion='entropy',max_depth=8)
DTC.fit(X_train,Y_train)
Y_pred_train = DTC.predict(X_train)
Y_pred_test = DTC.predict(X_test)
print('Number of nodes is',DTC.tree_.node_count)
print('Maximum depth is',DTC.tree_.max_depth)
print('Training score is',accuracy_score(Y_train,Y_pred_train))
print('Test score is',accuracy_score(Y_test,Y_pred_test))
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.4)
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
RFC = RandomForestClassifier(n_estimators=100,max_samples=0.6,max_features=0.8)
RFC.fit(X_train,Y_train)
Y_pred_train = RFC.predict(X_train)
Y_pred_test = RFC.predict(X_test)
print('Accuracy score for train is',accuracy_score(Y_train,Y_pred_train))
print('Accuracy score for test is',accuracy_score(Y_test,Y_pred_test))
```

```python
#applied standardization to your feature matrix X
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss_X = ss.fit_transform(X)
ss_X = pd.DataFrame(ss_X)
ss_X.columns = list(X)
ss_X
```

```python
#Implementing Leave-One-Out Cross-Validation (LOOCV) to evaluate the performance of your Random Forest Classifier (RFC). 
```

```python
from sklearn.model_selection import LeaveOneOut
import numpy as np
loo = LeaveOneOut()
train_error = []
test_error = []
for train_index,test_index in loo.split(df):
  X_train,X_test = ss_X.iloc[train_index],ss_X.iloc[test_index]
  Y_train,Y_test = Y[train_index],Y[test_index]
  RFC.fit(X_train,Y_train)
  Y_pred_train = RFC.predict(X_train)
  Y_pred_test = RFC.predict(X_test)
  train_error.append(accuracy_score(Y_train,Y_pred_train))
  test_error.append(accuracy_score(Y_test,Y_pred_test))
print('Accuracy score for train is',np.mean(train_error))
print('Accuracy score for test is',np.mean(test_error))
```

```python
#implemented k-fold cross-validation using KFold from scikit-learn to assess the performance of your Random Forest Classifier (RFC)
```

```python
from sklearn.model_selection import KFold
kfold = KFold(n_splits=6)
train_error = []
test_error = []
for train_index,test_index in kfold.split(df):
  X_train,X_test = ss_X.iloc[train_index],ss_X.iloc[test_index]
  Y_train,Y_test = Y[train_index],Y[test_index]
  RFC.fit(X_train,Y_train)
  Y_pred_train = RFC.predict(X_train)
  Y_pred_test  = RFC.predict(X_test)
  train_error.append(accuracy_score(Y_train,Y_pred_train))
  test_error.append(accuracy_score(Y_test,Y_pred_test))
print('Accuracy score for train is',np.mean(train_error))
print('Accuracy score for test is',np.mean(test_error))
```

```python
# Ensemble Methods
# Bagging Classifier
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
BAC = BaggingClassifier(base_estimator=DecisionTreeClassifier(),max_samples=0.6,max_features=0.7)
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
BAC.fit(X_train,Y_train)
```

```python
#BAC makes predictions on both the training set and the test set and evaluate the performance
```

```python
Y_pred_train = BAC.predict(X_train)
Y_pred_test = BAC.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('Training Accuracy is',accuracy_score(Y_train,Y_pred_train))
print('Test Accuracy is',accuracy_score(Y_test,Y_pred_test))
```

```python
# performing a grid search for hyperparameter tuning on your Random Forest Classifier (RFC)
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
params = [{'max_samples':np.arange(0.1,1.5,0.2),'max_features':np.arange(0.1,1.5,0.4)}]
GSV = GridSearchCV(BAC,params,cv=10,scoring='accuracy')
GSV.fit(X_train,Y_train)
```
