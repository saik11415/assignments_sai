# Delivery_time -> Predict delivery time using sorting time

```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('delivery_time.csv')
df.head()
df.shape
```

```python
X = df[['Sorting Time']]
X.head()
```

```python
Y = df['Delivery Time']
Y.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, Y, color='red')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X,Y)
```

```python
LR.intercept_
```

```python
LR.coef_
```

```python
Y_pred = LR.predict(X)
Y_pred
```

```python
plt.scatter(X, Y, color='red')
plt.scatter(X, Y_pred, color='blue')
plt.plot(X, Y_pred, color='violet')
plt.show()
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y, Y_pred)
print('mse is ',mse)
print('rmse is',np.sqrt(mse))
```

```python
X1 = X*X
LR.fit(X1,Y)
Y1_pred = LR.predict(X1)
plt.scatter(X1, Y, color='red')
plt.scatter(X1, Y1_pred, color='blue')
plt.plot(X1, Y1_pred, color='violet')
plt.show()
print('Intercept is',LR.intercept_)
print('Coef is',LR.coef_)
print('MSE is',mean_squared_error(Y,Y1_pred).round(2))
print('RMSE is',(np.sqrt(mean_squared_error(Y,Y1_pred))).round(2))
```

```python
X2 = np.sqrt(X)
LR.fit(X2,Y)
Y2_pred = LR.predict(X2)
plt.scatter(X2, Y, color='red')
plt.scatter(X2, Y2_pred, color='blue')
plt.plot(X2, Y2_pred, color='violet')
plt.show()
print('Intercept is',LR.intercept_)
print('Coef is',LR.coef_)
print('MSE is',mean_squared_error(Y,Y2_pred).round(2))
print('RMSE is',(np.sqrt(mean_squared_error(Y,Y2_pred))).round(2))
```
